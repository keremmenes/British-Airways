---
count: Bu, veri setindeki verilerin toplam sayısını temsil eder. Eksik değerler (NaN) dikkate alınmaksızın veri setindeki toplam giriş sayısını ifade eder.

unique: Bu, veri setinde benzersiz (farklı) değerlerin sayısını ifade eder. Yani, veri setinde kaç farklı değer olduğunu gösterir.

freq (frequency): Bu, veri setindeki en sık tekrar eden (en yaygın) değeri ve bu değerin kaç kez tekrarlandığını ifade eder. Yani, bu, en fazla tekrar eden değeri ve bu değerin sıklığını verir.

mean (ortalama): Bu, veri setindeki sayısal değerlerin aritmetik ortalamasını ifade eder. Tüm değerlerin toplamının veri noktalarının sayısına bölünmesiyle hesaplanır.

std (standart sapma): Bu, veri setindeki sayısal değerlerin dağılımının ne kadar değişken olduğunu ölçer. Düşük standart sapma, değerlerin ortalama etrafında daha yoğun bir şekilde toplandığını, yüksek standart sapma ise değerlerin ortalama etrafında daha yayıldığını gösterir.

min (minimum): Bu, veri setindeki en küçük değeri temsil eder.

25% (çeyrek 1): Bu, veri setinin alt çeyreğini temsil eder. Veriler küçükten büyüğe sıralandığında, en küçük değer ile 25. yüzde ilesi değeri arasındaki değeri ifade eder.

50% (medyan veya çeyrek 2): Bu, veri setinin ortanca değerini ifade eder. Veriler küçükten büyüğe sıralandığında, ortadaki değeri ifade eder. Medyan, verilerin dağılımına duyarlıdır ve aykırı değerlere karşı daha dirençlidir.

75% (çeyrek 3): Bu, veri setinin üst çeyreğini temsil eder. Veriler küçükten büyüğe sıralandığında, 75. yüzde ilesi değeri ile en büyük değer arasındaki değeri ifade eder.

max (maksimum): Bu, veri setindeki en büyük değeri temsil eder.

4.0 Data Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# kategorik değişkenleri sayısal değerlere dönüştürme

encode=OrdinalEncoder()

df[['sales_channel', 'trip_type', 'purchase_lead','flight_day' ,'route' , 'booking_origin']] = pd.DataFrame(encode.fit_transform(df[['sales_channel', 'trip_type', 'purchase_lead','flight_day' ,'route' , 'booking_origin']]))

#df[df['sales_channel'] == 1].head() sadece 1 olan değerlere bakmak için

df = df.astype('int64')
df.head()
```

```python
# Min-Max Normalizasyonu
scaler = MinMaxScaler()

columns_to_normalize = df.columns  # tüm sütunları seçmek için

df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

df.head()
```

```python
df.head(100)
df.describe(include='all')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_normalized = scaler.fit_transform(df)
df = pd.DataFrame(df_normalized, columns=df.columns)
df
```

```python
# Veriyi Eğitim ve Test Kümelerine Ayırma
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

df = shuffle(df, random_state=42)

y = df['booking_complete'] # Bağımlı değişken

#y = y.astype(int)

X=df.drop('booking_complete', axis=1) 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

df.head()
#y.unique()
```

5.0 Data Modelling

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)

model.fit(X_train, y_train)

feature_importance = model.feature_importances_
```

```python
feature_importance
```

6.0 Model Evaluation

```python
# Modeli Değerlendirme

# Çapraz Doğrulama (Cross-Validation)
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5)  # 5 kat çapraz doğrulama
```

```python
model.score(X_train,y_train)
```

```python
scores
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = model.predict(X_test)
```

```python
# Doğruluk (Accuracy) hesaplama
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# Hassasiyet (Precision) hesaplama
precision = precision_score(y_test, y_pred)
precision
```

```python
# Duyarlılık (Recall) hesaplama
recall = recall_score(y_test, y_pred)
recall
```

```python
#Karışıklık Matrisi (Confusion Matrix) hesaplama

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds", cbar=False, linewidths=1, linecolor="black")

plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Confusion Matrix')
class_labels = ["Negatif", "Pozitif"]
plt.xticks(ticks=[0.5, 1.5], labels=class_labels)
plt.yticks(ticks=[0.5, 1.5], labels=class_labels)

# Açıklamaları ekleyin
plt.text(0, 0, "TN", ha="right", va="center", color="black",rotation=0)
plt.text(0, 1, "FN", ha="right", va="center", color="black")
plt.text(1, 0, "FP", ha="left", va="top", color="black")
plt.text(1, 1, "TP", ha="left", va="top", color="black")


plt.show()
```
