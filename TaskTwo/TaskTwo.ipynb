4.0 Data Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import MinMaxScaler
```

```python
# kategorik deikenleri saysal deerlere dntrme

encode=OrdinalEncoder()

df[['sales_channel', 'trip_type', 'purchase_lead','flight_day' ,'route' , 'booking_origin']] = pd.DataFrame(encode.fit_transform(df[['sales_channel', 'trip_type', 'purchase_lead','flight_day' ,'route' , 'booking_origin']]))

#df[df['sales_channel'] == 1].head() sadece 1 olan deerlere bakmak iin

df = df.astype('int64')
df.head()
```

```python
# Min-Max Normalizasyonu
scaler = MinMaxScaler()

columns_to_normalize = df.columns  # tm stunlar semek iin

df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

df.head()
```

```python
df.head(100)
df.describe(include='all')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_normalized = scaler.fit_transform(df)
df = pd.DataFrame(df_normalized, columns=df.columns)
df
```

```python
# Veriyi Eitim ve Test Kmelerine Ayrma
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

df = shuffle(df, random_state=42)

y = df['booking_complete'] # Baml deiken

#y = y.astype(int)

X=df.drop('booking_complete', axis=1) 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

df.head()
#y.unique()
```

5.0 Data Modelling

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42)

model.fit(X_train, y_train)

feature_importance = model.feature_importances_
```

```python
feature_importance
```

6.0 Model Evaluation

```python
# Modeli Deerlendirme

# apraz Dorulama (Cross-Validation)
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=5)  # 5 kat apraz dorulama
```

```python
model.score(X_train,y_train)
```

```python
scores
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = model.predict(X_test)
```

```python
# Doruluk (Accuracy) hesaplama
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# Hassasiyet (Precision) hesaplama
precision = precision_score(y_test, y_pred)
precision
```

```python
# Duyarllk (Recall) hesaplama
recall = recall_score(y_test, y_pred)
recall
```

```python
#Karklk Matrisi (Confusion Matrix) hesaplama

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds", cbar=False, linewidths=1, linecolor="black")

plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Confusion Matrix')
class_labels = ["Negatif", "Pozitif"]
plt.xticks(ticks=[0.5, 1.5], labels=class_labels)
plt.yticks(ticks=[0.5, 1.5], labels=class_labels)

# Aklamalar ekleyin
plt.text(0, 0, "TN", ha="right", va="center", color="black",rotation=0)
plt.text(0, 1, "FN", ha="right", va="center", color="black")
plt.text(1, 0, "FP", ha="left", va="top", color="black")
plt.text(1, 1, "TP", ha="left", va="top", color="black")


plt.show()
```
